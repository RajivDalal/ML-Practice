# Pandas
- Library that helps to manipulate and analyze data
- Provides data frames and series

```python
import numpy as np
import pandas as pd
```

# Series


```python
data = ["Rajiv",20,"ab",2004]
s=pd.Series(data)
s
```

```python
s = pd.Series(np.random.rand(15))
print(s)
print(s[3])
```

```python
print(s[0:2])
```

```python
print(s.argmax())
print(s.argmin())
print(s.mean())
print(s.max())
print(s.min())
print(s.size) #s.count() also same
```

```python
print(s.head()) #Top 5 records
print(s.tail()) #Bottom 5 records
```

```python
print(s.describe())
```

# Dataframes

```python
s1 = pd.Series(np.random.rand(5))
s2 = pd.Series(np.random.rand(5))
print(s1,s2,sep="\n")
```

```python
#Creating dataframes using series
a = pd.DataFrame((s1,s2))
print(a)
```

```python
#Creating dataframes using dictionaries
d1 = {"name":"Rajiv","age":20}
d2 = {"name":"Srishti","age":21}

b = pd.DataFrame((d1,d2))
print(b)
```

```python
b["gender"]=["Male","Female"]
print(b)
```

#iloc & loc functions
- dataframe_name.iloc[row_index,column_index]
- dataframe_name.iloc[row_name,column_name]

```python
print(b.iloc[0,1])
print(b.iloc[0:2,1:])
```

```python
print(b.loc[:,["name","gender"]])
```

```python
print(b.shape)
print(b.columns)
```

```python
print(b.isnull(),end="\n\n")
print(b.isnull().sum())
```

```python
b.rename(columns={"exp":"age"},inplace=True)
print(b)
```

```python
#Dropping a column
b.drop(columns=["gender"],axis=1) #axis=0 for rows
```

```python
b.duplicated()
```

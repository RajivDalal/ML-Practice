# Seaborn

```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
data = sns.load_dataset("tips")
data
```

```python
data.info()
```

# Countplot

```python
sns.countplot(data=data,x="smoker",palette=["r","g"])
```

# Distplot

```python
sns.distplot(data.total_bill)
```

```python
sns.distplot(data.total_bill,hist=False)
```

```python
sns.scatterplot(data=data,x="total_bill",y="tip", hue="sex",size="smoker")
```

# Joint Plot

```python
sns.jointplot(data=data,x="total_bill",y="tip")
```

# Boxplot

```python
sns.boxplot(data=data,x="total_bill")
```

# Pair plot

```python
sns.pairplot(data)
```

# Correlation
- Numeric
- Heatmap

```python
hm = data.corr(numeric_only=True)
hm
```

```python
sns.heatmap(hm,annot=True)
```

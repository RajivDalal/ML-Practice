# Matplotlib

```python
import matplotlib.pyplot as plt
```

# Line Plot

```python
x = [2010,2011,2012,2013,2014]
y = [5,7,6,2,4]
plt.plot(x,y)
```

```python
plt.plot(x,y,marker="o",c="r",mfc="b",ls="--")
```

```python
x=[1,2,3,4,5]
y=[1,2,3,4,5]
plt.plot(x,y)
plt.title("Line Plot")
plt.xlabel("X Label")
plt.ylabel("Y Label")
```

# Bar Chart

```python
x = ["Toys","Grocery","Books","Shoes"]
y = [9,8,3,5]
plt.bar(x,y)
```

```python
plt.barh(x,y, color=["r","gold","g","b"])
```

# Histogram

```python
x=[9,4,6,7,8,9,9,2,4,4,6,6,7,6]
plt.hist(x)
```

# Scatter Plots

```python
x=[6,2,4,5,6,7,3,4,5,4]
y=[9,3,5,5,2,4,5,7,9,4]
plt.scatter(x,y)
```

# Pie Chart

```python
x = [200, 150, 250, 375, 240]
mylabels = ["Swift","Dzire","WagonR","Mercedes","Alto"]
plt.pie(x,labels=mylabels,autopct="%.2f%%",explode=[0,0,0.2,0,0])
plt.legend()
plt.show()
```

# Working with csv file

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("/content/insurance.csv")
data
```

```python
data.head()
```

```python
plt.hist(data["age"])
```

```python
plt.bar(data["age"],data["bmi"])
```

```python
x = data["sex"].unique()
y = data["sex"].value_counts()

plt.bar(x,y)
```

```python
y = data["region"].value_counts()
print(y)
x = y.index
plt.bar(x,y)
```

```python
plt.pie(y,labels=x,autopct="%.2f%%")
```

# Subplots
Three arguments
- Row Value
- Column value
- Index of current plot

```python
plt.subplot(2,2,1)
plt.subplot(2,2,2)
plt.subplot(2,2,3)
plt.subplot(2,2,4)
```

```python
plt.figure(figsize=(10,6))

plt.subplot(2,2,1)
plt.bar(data["age"],data["bmi"])

plt.subplot(2,2,2)
plt.bar(data["age"],data["smoker"])

plt.subplot(2,2,3)
plt.bar(data["age"],data["sex"])

plt.subplot(2,2,4)
plt.bar(data["age"],data["children"])
```

# Figure Axis method


```python
fig,ax=plt.subplots(2,2)
ax[0,0].plot(range(10),"r")
ax[0,1].plot(range(10),"b")
ax[1,0].plot(range(10),"g")
ax[1,1].plot(range(10),"y")
```
